# Analisando o Uso de Medicamentos
* Esta tarefa analisará os dados conforme três abordagens: análise do comportamento estatístico; análise de correlação entre pares; análise de correlação de variáveis.
* A partir dos arquivos CSV que estão no diretórios /data/nhanes considere as seguintes tabelas SQL.

```sql
DROP Table IF EXISTS Person;
DROP Table IF EXISTS Drug;
DROP Table IF EXISTS DrugUse;
```

```sql
CREATE TABLE Person (
  Id VARCHAR(5) NOT NULL,
  Gender INTEGER,
  Age INTEGER,
  PRIMARY KEY(Id)
) AS SELECT
    id,
    gender,
    age
FROM CSVREAD('../../../data/nhanes/demographic-person.csv');

CREATE TABLE Drug (
  Code VARCHAR(6) NOT NULL,
  Name VARCHAR(100) NOT NULL,
  PRIMARY KEY(Code)
) AS SELECT
    code,
    name
FROM CSVREAD('../../../data/nhanes/medications-drug.csv');

CREATE TABLE DrugUse (
  PersonId VARCHAR(5) NOT NULL,
  DrugCode VARCHAR(6) NOT NULL,
  DaysUse INTEGER,
  FOREIGN KEY(PersonId)
    REFERENCES Person(Id)
      ON DELETE NO ACTION
      ON UPDATE NO ACTION,
  FOREIGN KEY(DrugCode)
    REFERENCES Drug(Code)
      ON DELETE NO ACTION
      ON UPDATE NO ACTION
) AS SELECT
    person_id,
    drug_code,
    days_use
FROM CSVREAD('../../../data/nhanes/medications-use.csv');
```

```sql
SELECT * FROM Person;
SELECT * FROM Drug;
SELECT * FROM DrugUse
```

# I. Análise de comportamento estatístico

## 1) Contabilizando o uso de medicamentos

* Liste o nome de todos os medicamentos e a quantidade de pessoas que usa cada um deles.

```sql
SELECT D.Name, COUNT(*) FROM DrugUse M, Drug D WHERE M.DrugCode = D.Code GROUP BY M.DrugCode 
```

## 2) Medicamento mais usado com VIEW

* Informe o código do medicamento mais usado: fazendo uso de VIEW.

```sql
CREATE VIEW NumberDrugs AS SELECT DrugCode, COUNT(*) Quantity FROM DrugUse GROUP BY DrugCode
```

```sql
SELECT DrugCode FROM NumberDrugs WHERE  Quantity = (SELECT MAX(Quantity) FROM NumberDrugs)
```

## 3) Medicamento mais usado sem VIEW

* Informe o código do medicamento mais usado: sem fazer uso de VIEW.

```sql

SELECT DrugCode FROM (SELECT DrugCode, COUNT(*) Quantity FROM DrugUse GROUP BY DrugCode) 
                WHERE  Quantity = (SELECT MAX(Quantity) 
                FROM (SELECT DrugCode, COUNT(*) Quantity FROM DrugUse GROUP BY DrugCode))
```

## 4) Nome do medicamento mais usado
* Informe o nome do medicamento mais usado (uso de VIEW é opcional).

```sql
SELECT Name FROM Drug 
            WHERE Code = (SELECT DrugCode 
                          FROM(SELECT DrugCode, COUNT(*) Quantity FROM DrugUse GROUP BY DrugCode) 
                          WHERE  Quantity = (SELECT MAX(Quantity) FROM (SELECT DrugCode, COUNT(*) Quantity FROM DrugUse GROUP BY DrugCode)))
```

## 5) Contabilizando quanto as pessoas usam de medicamento

* Informe o quantidade média de medicamentos diferentes usados por pessoa.

```sql
DROP VIEW IF EXISTS PersonDrugs;
CREATE VIEW PersonDrugs AS
    SELECT PersonId, COUNT(PersonId) QuantityDrugs
    FROM DrugUse
    GROUP BY(PersonId);
    
SELECT AVG(QuantityDrugs) FROM PersonDrugs;
```

## 6) Pessoas que usam mais do que a média

* Liste o id das pessoas que usam mais medicamentos do que a média.

```sql
DROP VIEW IF EXISTS PersonDrugs;
CREATE VIEW PersonDrugs AS
    SELECT PersonId, COUNT(PersonId) QuantityDrugs
    FROM DrugUse
    GROUP BY(PersonId);
    
SELECT PersonId FROM PersonDrugs WHERE QuantityDrugs > (SELECT AVG(QuantityDrugs) FROM PersonDrugs)
```

## 7) Análise do uso de medicamento

* Considere que um médico quer fazer responder a seguinte questão: pessoas tendem a usar mais medicamentos conforme ficam mais velhas?
* Escreva uma query que ajude o médico a realizar esta análise.

```sql
DROP VIEW IF EXISTS PersonDrugs;
CREATE VIEW PersonDrugs AS
    SELECT PersonId, COUNT(PersonId) QuantityDrugs
    FROM DrugUse
    GROUP BY(PersonId);
    
    
SELECT AVG(PD.QuantityDrugs) FROM PersonDrugs PD WHERE PD.PersonId IN (SELECT Id FROM Person WHERE Age < 20);

SELECT AVG(PD.QuantityDrugs) FROM PersonDrugs PD WHERE PD.PersonId IN (SELECT Id FROM Person WHERE Age < 40 AND Age > 20);

SELECT AVG(PD.QuantityDrugs) FROM PersonDrugs PD WHERE PD.PersonId IN (SELECT Id FROM Person WHERE Age < 60 AND Age > 40);

SELECT AVG(PD.QuantityDrugs) FROM PersonDrugs PD WHERE PD.PersonId IN (SELECT Id FROM Person WHERE Age < 80 AND Age > 60);
```

```sql
DROP TABLE IF EXISTS Taxonomia;
DROP TABLE IF EXISTS Marcadores;

CREATE TABLE Taxonomia (
  Categoria VARCHAR(50) NOT NULL,
  Superior  VARCHAR(50),
  PRIMARY KEY(Categoria),
  FOREIGN KEY(Superior)
    REFERENCES Taxonomia(Categoria)
      ON DELETE NO ACTION
      ON UPDATE CASCADE
);

CREATE TABLE Marcadores (
  Titulo VARCHAR(50) NOT NULL,
  Endereco VARCHAR(80) NOT NULL,
  Acessos INTEGER,
  Categoria VARCHAR(50) NOT NULL,
  PRIMARY KEY(Titulo),
  FOREIGN KEY(Categoria)
    REFERENCES Taxonomia(Categoria)
      ON DELETE NO ACTION
      ON UPDATE CASCADE
);
```
